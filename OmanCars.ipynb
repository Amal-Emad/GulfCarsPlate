```python
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

input_image_path = '//content/drive/MyDrive/car&plates/omancars_sample'
output_dir2 = '/content/drive/MyDrive/CAR&PLATE/output_dir2'
model_path = '/content/drive/MyDrive/car&plates/vehicle_yolov8s.pt'

if not os.path.exists(output_dir2):
    os.makedirs(output_dir2)

vehicle_model = YOLO(model_path)
results = vehicle_model.predict(input_image_path, save_txt=True, save_crop=True, project=output_dir2)


for result in results:
    if 'crops' in result:
        for crop in result['crops']:
            crop_path = os.path.join(output_dir2, f'crop_{crop["id"]}.jpg')
            cv2.imwrite(crop_path, crop['im'])

            img = cv2.cvtColor(crop['im'], cv2.COLOR_BGR2RGB)
            plt.figure()
            plt.title(f'Crop {crop["id"]}')
            plt.axis('off')
            plt.imshow(img)
            plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os

directory = '/content/drive/MyDrive/CAR&PLATE/output_dir2/predict4/crops/vehicle'
files = [file for file in os.listdir(directory) if file.endswith(('.jpg', '.png'))]


files = files[:9]


cols = 3
rows = 3

plt.figure(figsize=(15, 10))

for i, file_name in enumerate(files):
    image_path = os.path.join(directory, file_name)
    image = cv2.imread(image_path)

    if image is not None:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.subplot(rows, cols, i + 1)
        plt.imshow(image_rgb)
        plt.axis('off')
        plt.title(file_name)

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os
import math

directory = '/content/drive/MyDrive/CAR&PLATE/car&plate_code/output_plate/predict/crops/Square Plate'
files = [file for file in os.listdir(directory) if file.endswith(('.jpg', '.png'))]

num_images = len(files)


rows = math.ceil(num_images / cols)


plt.figure(figsize=(15, 10))

for i, file_name in enumerate(files):
    image_path = os.path.join(directory, file_name)
    image = cv2.imread(image_path)

    if image is not None:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.subplot(rows, cols, i + 1)
        plt.imshow(image_rgb)
        plt.axis('off')
        plt.title(file_name)

plt.tight_layout()
plt.show()
```
